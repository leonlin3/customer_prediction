```python
def select_complete_variable(data, target, n):
    corr = np.abs(data.corrwith(target)).sort_values()                       # 计算相关度，只考虑相关性大小，不考虑正负
    # 判断某个变量是否为完全变量
    for i in list(corr.index):
        if sum(data[i] == -1) != 0 or sum(data[i] == -10000) != 0 or sum(data[i].isnull()) != 0:
            corr.drop(index=i, inplace=True)                                 # 若有空值则不为完全变量
    return list(corr[-n:].index)

complete_variable = select_complete_variable(numeric_feature, target, 3)                # 选择3个聚类的完全变量
_X = numeric_feature[complete_variable]      
```

```python
def K_means(x, n):
    estimator = KMeans(n_clusters=n)
    estimator.fit(x)
    return estimator.labels_
data_labels = K_means(_X, 5)                                                  # 每个样本的类别标签
numeric_feature['label'] = data_labels                                                   # 新增一列为类别标签
```

```python
numeric_feature.drop(['l12_mon_insu_buy_whl_tms','l12_mon_gld_buy_whl_tms', 'ovd_30d_loan_tot_cnt'], axis = 1, inplace = True)
numeric_feature['cur_debit_cnt'] += 1e-4
numeric_feature['cur_credit_cnt'] += 1e-4
numeric_feature['cur_debit_min_opn_dt_cnt'] = numeric_feature['cur_debit_min_opn_dt_cnt'] / numeric_feature['cur_debit_cnt']
numeric_feature['cur_credit_min_opn_dt_cnt'] = numeric_feature['cur_credit_min_opn_dt_cnt'] / numeric_feature['cur_credit_cnt']
numeric_feature.drop(['cur_debit_cnt', 'cur_credit_cnt'], axis = 1, inplace = True)
```

```python
numeric_feature['Trx_Cod2_Cd'].fillna(-10000, inplace = True)
numeric_feature['trx/Flg1_B'].fillna(-10000, inplace = True)
numeric_feature['trx/Flg1_C'].fillna(-10000, inplace = True)
numeric_feature['Trx_Cod1_Cd'].fillna(-10000, inplace = True)
numeric_feature['cny_trx_amt_thisMonth'].fillna(-10000, inplace = True)
numeric_feature['trd_count'].fillna(-10000, inplace = True)
numeric_feature['page_no'].fillna(-10000, inplace = True)
# 根据类别标签对特征空值进行填充
for i in range(5):
    for j in list(numeric_feature.columns):
        tmp = numeric_feature.loc[numeric_feature.label == i, j]
        mean = np.mean(tmp.loc[(tmp != -1) & (tmp != -10000)])
        if sum(numeric_feature[j] == -1):
            numeric_feature.loc[(numeric_feature[j] == -1) & (numeric_feature.label == i), j] = mean
        if sum(numeric_feature[j] == -10000):
            numeric_feature.loc[(numeric_feature[j] == -10000) & (numeric_feature.label == i), j] = mean
```
